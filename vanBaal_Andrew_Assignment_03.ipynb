Write a program that reads in a list of exam scores and assigns grades to each student based on the following scale: A for scores between 90 and 100, B for scores between 80 and 89, C for scores between 70 and 79, D for scores between 60 and 69, and F for scores below 60. Use a for loop and an if/elif statement to implement the grading logic. **(3 points)**

```python
scores = [95, 76, 82, 99, 65, 25]
```

```python
letter_score = []
for score in scores:
  if score > 90:
    letter_score.append('A')
  elif 80<score<89:
    letter_score.append('B')
  elif 70<score<79:
    letter_score.append('C')
  elif 60<score<69:
    letter_score.append('D')
  elif score < 60:
    letter_score.append('F')

for score in range(len(scores)):
    print(f"Student {score+1}: {scores[score]},{letter_score[score]} ")
```

Write a program that uses a while loop to prompt the user to enter a number between 18 and 58. If the user enters a number outside of this range, the program should keep prompting them to enter a valid number. Once the user enters a valid number, the program should print ``Your Input: [number]". **(3 points)**

```python
# Creating input storage variable
user_input = 0

# creating while loop for conditional input
while user_input < 18 or user_input > 58:
    user_input = int(input("Enter a number between 18 and 58: "))
    if user_input < 18 or user_input > 58:
        print("Invalid. Please enter a number between 18 and 58.")

# printing user's result
print("Your Input:", user_input)
```

Write a program that reads ``sample.csv" file containing two columns of data (xData and yData) and fits a linear regression model to the data using the statsmodels.api.OLS class. The program should print the slope and intercept of the regression line, as well as the R-squared value. **(5 points)**

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# reading in the dataframe
df = pd.read_csv('sample.csv')
```

```python
# assigning x and y arrays for the regression model
x = df['xData']
y = df['yData']

# creating the initial scatterplot
plt.scatter(x, y)
plt.title('Scatter Plot of X and Y Data')
plt.xlabel('X Data')
plt.ylabel('Y Data')

# adding constant to the x array for the intercept
x = sm.add_constant(x)

# fitting the model
model = sm.OLS(y, x).fit()

# printing desired parameters
# print(model.summary()) this will print the entire regression summary if desired
print('Slope: ', model.params[1])
print('Intercept: ', model.params[0])
print('R-squared: ', model.rsquared)

# assigning slope and intercept to fit line
slope = model.params[1]
intercept = model.params[0]

x_line = np.linspace(x.min(), x.max(), 100)

y_line = intercept + slope * x_line

# plotting the regression line atop the existing scatterplot
plt.plot(x_line, y_line, color = 'red')
plt.show()
```
